```python
x = np.arange(-2*np.pi, 2*np.pi, .1)
y = np.sin(x)
plt.title('Sin Curve')
plt.plot(x, y)
```

# Subplots

### With axes shift and custom ticks

```python
p1 = plt.subplot(2, 2, 1)
p1.spines['bottom'].set_position(('data', 0))
p1.spines['left'].set_position(('data', 0))
p1.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.sin(x), 'r')

p2 = plt.subplot(2, 2, 2)
p2.spines['bottom'].set_position(('data', 0))
p2.spines['left'].set_position(('data', 0))
p2.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.cos(x), 'g')

p3 = plt.subplot(2, 2, 3)
p3.spines['bottom'].set_position(('data', 0))
p3.spines['left'].set_position(('data', 0))
p3.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.tan(x))
plt.ylim(-10, 10)
```

# Creating a figure and adding subplots

```python
x = np.arange(-10, 10)
y = x**2
z = x**3
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.5,0.15,.4,.3])

ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('z = x^3')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('y = x^2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, color = "blue", lw = 3, ls = '--')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = x^2')

axes[1].scatter(x, z, color = "red", lw = 3)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('z = x^3')
fig.tight_layout()
```

# Other Modifications

#### Logarithmic scale 
#### Fill between
#### Annotations
#### Axis grid
#### Axis spines

```python
x = np.arange(6)
fig, axes = plt.subplots(1, 2, figsize = (10,4))
      
axes[0].fill_between(x, x**2, np.exp(x), color = 'green', alpha = .5)
axes[0].set_title("Normal scale")
axes[0].text(3.2, 100, r"$y = log(x)$", fontsize = 12, color = 'red')
axes[0].text(4.2, 10, r"$y = x^2$", fontsize = 12, color = 'red')
axes[0].grid(color = 'green', ls = 'dashed')
axes[0].spines['bottom'].set_color('red')
axes[0].spines['bottom'].set_linewidth(2)
axes[0].spines['left'].set_color('red')
axes[0].spines['left'].set_linewidth(2)

axes[1].fill_between(x, x**2, np.exp(x), alpha = .5)
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
axes[1].text(3, 80, r"$y = log(x)$", fontsize = 12, color = 'red')
axes[1].text(4, 12, r"$y = x^2$", fontsize = 12, color = 'red')
axes[1].grid(color = 'blue', ls = 'dashed')
axes[1].spines['right'].set_color('red')
axes[1].spines['right'].set_linewidth(2)
axes[1].yaxis.tick_right()
axes[1].spines['bottom'].set_color('red')
axes[1].spines['bottom'].set_linewidth(2)
```

# us-marriages-divorces-1867-2014 dataset

#### Create a line plot showing the number of marriages and divorces in the U.S. between 1867 and 2014. Label both of the lines and show the legend.

```python
us_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
```

```python
year = us_data['Year'].values
marriages = us_data['Marriages'].values
divorces = us_data['Divorces'].values

plt.plot(year, marriages, label = 'Marriages')
plt.plot(year, divorces, label = 'Divorces')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
```

#### Create a vertical bar chart comparing the number of marriages and divorces in the U.S. between the years 1991 and 2000.

```python
newdata=us_data.dropna()[(us_data.Year>1991)&(us_data.Year<2000)]
newdata.plot(x="Year", y=["Marriages", "Divorces"], kind="bar")
```

#### Use subplots to create a figure with two separate line plots showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Place the two line plots on top of each other so the years line up.

```python
fig, (ax1, ax2) = plt.subplots(2, sharex = 'col', sharey = 'row')
ax1.plot(us_data.Year, us_data.Marriages_per_1000, color = 'red')
ax1.set(ylabel = 'Count per 1000', title = 'Marriages per 1000')
ax2.plot(us_data.Year, us_data.Divorces_per_1000)
ax2.set(xlabel = 'Year', ylabel = 'Count per 1000',title = 'Divorces per 1000')
```

# actor_kill_counts dataset

#### Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.

```python
actors = pd.read_csv('actor_kill_counts.csv')
name = actors['Actor'].values
kills = actors['Count'].values
plt.barh(name, kills)
plt.xlabel('Kills')
plt.ylabel('Actor')
```

# recent-college-grads-earnings dataset

#### Create two box plots to compare the earnings of recent college graduates in majors that are primarily composed of women to majors that are primarily composed of men.

```python
df = pd.read_csv('recent-college-grads-earnings.csv')
Q1_W = df.loc[df["ShareWomen"]>0.5, "Median"]
Q1_M = df.loc[df["ShareWomen"]<=0.5, "Median"]
Q1 = [Q1_W, Q1_M]
labels = ("Women", "Men")
positions = (1, 2)
plt.boxplot(Q1, patch_artist = True)
plt.xticks(positions, labels)
plt.ylabel('Earnings')
plt.title('Earnings of Recent College Graduates in\nMajors primarily composed of Women vs Majors primarily composed of Men')
```

# percent-degrees-conferred-women-usa dataset

#### Use small multiples to compare the trends in gender ratio of 17 U.S. college majors between 1970 and 2011.

```python
df = pd.read_csv('percent-degrees-conferred-women-usa.csv')
fig = plt.figure(figsize = (20, 18))
c = 1
for i in df.columns[1:]:
  plt.subplot(5, 4, c)
  plt.plot(df.Year, df[i])
  plt.title('{}'.format(i))
  plt.xlabel('Year')
  c = c + 1
plt.subplots_adjust(bottom=15, right=15, top=20, wspace=10, hspace=10)
fig.tight_layout(pad=6.0,)
fig.suptitle('Percentage of Women in U.S. college majors between 1970 and 2011')
```
