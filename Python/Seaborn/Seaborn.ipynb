```python
import pandas as pd
import seaborn as sns
```

```python
df =pd.read_csv("H:/Jupyter Notebooks/insurance.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.corr()
```

# Heatmap

```python
sns.heatmap(df.corr())
```

# Joint Plot

```python
sns.jointplot(x = 'bmi', y = 'charges', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'age', y = 'charges', data = df, hue = 'smoker')
```

# Pair Plot

```python
sns.pairplot(data = df, kind = 'reg', diag_kws = dict(kde = True, edgecolor = "none"))
```

# Dist Plot

```python
sns.distplot(df['charges'])
```

# Count Plot

```python
sns.countplot('smoker', data = df)
```

# Bar Plot

```python
sns.barplot(y = 'sex', x = 'charges', data = df)
```

# Box Plot

```python
sns.boxplot(x = 'region', y = 'charges', data = df, palette = 'rainbow')
```

```python
sns.boxplot(y = 'region', x = 'charges', data = df, hue = 'sex')
```

# Violin Plot

```python
sns.violinplot(x = 'region', y = 'charges', data = df, palette = 'rainbow')
```
