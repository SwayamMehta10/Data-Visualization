```python
import numpy as np
```

```python
x = np.arange(0, 20)
y = np.arange(21, 41)
```

# Simple Plots

```python
plt.scatter(x, y ,c = 'b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Scatterplot')
# plt.savefig('Test.png') to save plot as png.
```

```python
y = x**3
plt.plot(x, y, color = 'r', marker = 'D', linestyle = 'dashed', linewidth = 2, markersize = 5)
# plt.plot(x, y, 'ro--', lw= 2, ms = 5)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('y = x^3')
```

```python
x = np.arange(-2*np.pi, 2*np.pi, .1)
y = np.sin(x)
plt.title('Sin Curve')
plt.plot(x, y)
```

# Subplots

### With axes shift and custom ticks

```python
p1 = plt.subplot(2, 2, 1)
p1.spines['bottom'].set_position(('data', 0))
p1.spines['left'].set_position(('data', 0))
p1.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.sin(x), 'r')

p2 = plt.subplot(2, 2, 2)
p2.spines['bottom'].set_position(('data', 0))
p2.spines['left'].set_position(('data', 0))
p2.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.cos(x), 'g')

p3 = plt.subplot(2, 2, 3)
p3.spines['bottom'].set_position(('data', 0))
p3.spines['left'].set_position(('data', 0))
p3.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
plt.plot(x, np.tan(x))
plt.ylim(-10, 10)
```

# Creating a figure and adding subplots

```python
x = np.arange(-10, 10)
y = x**2
z = x**3
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.5,0.15,.4,.3])

ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('z = x^3')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('y = x^2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, color = "blue", lw = 3, ls = '--')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = x^2')

axes[1].scatter(x, z, color = "red", lw = 3)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('z = x^3')
fig.tight_layout()
```

# Barplot

```python
x1 = np.random.randint(20, size = 5)
x2 = np.random.randint(20, size = 5)
y1 = np.random.randint(20, size = 5)
y2 = np.random.randint(20, size = 5)
print('x1 =', x1)
print('y1 =', y1)
print('x2 =', x2)
print('y2 =', y2)
plt.bar(x1, y1, alpha = .75)
plt.bar(x2, y2, alpha = .75, color = 'g')
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

# Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 6)]
plt.boxplot(data, patch_artist = True); 
```

# Pie Chart

```python
labels = 'Python', 'C', 'C++', 'Java', 'R'
sizes = [245, 190, 180, 135, 155]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
explode = (.1, 0, 0, 0, 0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
```

# Other Modifications

#### Logarithmic scale 
#### Fill between
#### Annotations
#### Axis grid
#### Axis spines

```python
x = np.arange(6)
fig, axes = plt.subplots(1, 2, figsize = (10,4))
      
axes[0].fill_between(x, x**2, np.exp(x), color = 'green', alpha = .5)
axes[0].set_title("Normal scale")
axes[0].text(3.2, 100, r"$y = log(x)$", fontsize = 12, color = 'red')
axes[0].text(4.2, 10, r"$y = x^2$", fontsize = 12, color = 'red')
axes[0].grid(color = 'green', ls = 'dashed')
axes[0].spines['bottom'].set_color('red')
axes[0].spines['bottom'].set_linewidth(2)
axes[0].spines['left'].set_color('red')
axes[0].spines['left'].set_linewidth(2)

axes[1].fill_between(x, x**2, np.exp(x), alpha = .5)
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
axes[1].text(3, 80, r"$y = log(x)$", fontsize = 12, color = 'red')
axes[1].text(4, 12, r"$y = x^2$", fontsize = 12, color = 'red')
axes[1].grid(color = 'blue', ls = 'dashed')
axes[1].spines['right'].set_color('red')
axes[1].spines['right'].set_linewidth(2)
axes[1].yaxis.tick_right()
axes[1].spines['bottom'].set_color('red')
axes[1].spines['bottom'].set_linewidth(2)
```
